```python
import os
import pandas as pd
import ast
import json

# Load the csv file containing all the Le Louvre artworks and corresponding information
current_dir = os.getcwd()
csv_path = os.path.join(current_dir, '..', '2_retrieve_all_infos_all_works_of_art_using_API', 'all_works_of_art_le_louvre_uncleaned.csv')

# Read the CSV file
df = pd.read_csv(csv_path, index_col=0)

# Display the dataframe
df
```

```python
# Display the name of the columns of the dataframe
df.columns
```

```python
# Display the unique values of the column
df["dateCreated"].unique()
```

```python
# Function to extract startYear and endYear
def extract_years(date_str):
    try:
        date_list = ast.literal_eval(date_str)
        start_year = date_list[0]['startYear']
        end_year = date_list[0]['endYear']
        return start_year, end_year
    except (ValueError, KeyError, IndexError):
        return None, None

# Use the function to create 2 new columns, startYear and endYear
df['startYear'], df['endYear'] = zip(*df['dateCreated'].apply(extract_years))

df[["dateCreated", "startYear", "endYear"]]
```

```python
# Function to extract the name of the artwork creator (if existing)
def extract_creator(json_string):
    try:
        json_data = json.loads(json_string.replace("'", '"'))
        if json_data:
            return json_data[0].get('label', None)
        else:
            return None
    except:
        return None
```

```python
# Use the function to create a new column "creatorName" with the creator's name
df['creatorName'] = df['creator'].apply(extract_creator)

# Display the list of unique values of the "creatorName" column
df['creatorName'].unique()
```

```python
# Display the unique values of the column "collection"
df["collection"].unique()
```

```python
# Display the unique values of the column "image"
df["image"].unique()
```

```python
import re

# function to extract the url of the image

def extract_url_image(string):
    # Utiliser une expression régulière pour rechercher la sous-chaîne entre "'urlImage': '" et "''"
    match = re.search(r"'urlImage': '(.+?)'", string)
    if match:
        return match.group(1)
    else:
        return None

```

```python
# Apply the function to the "image" column
df['image'] = df['image'].apply(extract_url_image)
df["image"]
```

```python
# Create a dataframe with only the main information about the artworks
simplified_le_louvre_works_of_art = df[[
                                        "title",
                                        #"titleComplement",
                                        "collection",
                                        "room",
                                        'startYear',
                                        'endYear',
                                        #"index.Period",
                                        'creatorName',
                                        #"index.Materials",
                                        #"ownedBy",
                                        #"index.Mode d\'acquisition",
                                        "url",
                                        "image"
                                        ]]

# Display the dataframe
simplified_le_louvre_works_of_art
```

```python
# Check how many artworks have a 'room' value that is not missing in the corresponding column :
# we suppose that this is how we know which artworks are currently displayed
simplified_le_louvre_works_of_art[simplified_le_louvre_works_of_art['room'].notna()]
```

```python
# Save the dataframe with main information about the artworks currently on display in a csv file
simplified_le_louvre_works_of_art[simplified_le_louvre_works_of_art['room'].notna()].to_csv('simplified_le_louvre_works_of_art_on_display_all_dpts.csv')
```

```python
# Count the number of artworks on display by collection
simplified_le_louvre_works_of_art[simplified_le_louvre_works_of_art['room'].notna()]['collection'].value_counts()
```

```python
# Create a list containing the name of the collection linked to the Antiquity 
collections_cibles = [
    "Département des Antiquités grecques, étrusques et romaines",
    "Département des Antiquités orientales",
    "Département des Antiquités égyptiennes"
]

# Display only the artworks on display and corresponding to the collections mentioned above
antiquities_df = simplified_le_louvre_works_of_art[
    (simplified_le_louvre_works_of_art['room'].notna()) &  
    (simplified_le_louvre_works_of_art['collection'].isin(collections_cibles))  
]

# Display the results
antiquities_df
```

```python
# Save the dataframe of the artworks on display linked to Antiquity in a csv file
antiquities_df.to_csv("simplified_le_louvre_works_of_art_on_display_antiquities.csv", index=False)
```

```python
# Save the dataframe of the artworks on display linked to Middle Ages, Renaissance and Modern Times in a csv file
middle_ages_renaissance_modern_times_df.to_csv("simplified_le_louvre_works_of_art_on_display_ma_ren_mt_sculptobjects.csv", index=False)
```
